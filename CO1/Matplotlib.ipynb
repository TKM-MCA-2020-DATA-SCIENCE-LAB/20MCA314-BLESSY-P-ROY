matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x ** 2
```

```python
y
```

```python
x
```

basic matplotlib command

```python
plt.plot(x,y,'g')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('String title')
plt.show()
```

create multiplots in same fucntion

```python
plt.subplot(7,3,1)
plt.plot(x,y,'g--')
plt.subplot(1,3,2)
plt.plot(y,x,'b*-');
plt.subplot(2,3,3)
plt.plot(x,y,'r')
```

using object oriented method

```python
fig=plt.figure()
axes=fig.add_axes([0.5,0.1,0.7,0.8])
axes.plot(x,y,'r--')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

subplot

```python
fig=plt.figure()
axes1=fig.add_axes([0.2,0.1,0.9,0.9])
axes2=fig.add_axes([0.4,0.5,0.3,0.4])
axes1.plot(x,y,'r--')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('Axis1')
axes2.plot(y,x,'g')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Axis2')
```

```python
fig=plt.figure(figsize=(9,4),dpi=100)

```

```python
fig,axes=plt.subplots(2,2,figsize=(6,7))
```

```python
for ax in axes:
  ax.plot(x,y,'r--')
  ax.set_xlabel('X')
  ax.set_ylabel('Y')
  ax.set_title('Title');
```

legend

```python
fig, ax = plt.subplots()

ax.plot(x, x**2, label="y = x**2")
ax.plot(x, x**3, label="y = x**3")
ax.legend(loc=2); # upper left corner
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title');
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(1,25,1)
y=np.log(x)
plt.plot(x,y,marker='x')
plt.show()
```

```python
plt.plot(x,y,marker='D')
plt.title('Sin Function')
plt.xlabel('x values')
plt.ylabel('y label')
plt.show()
```
