Seaborn Distribution Plots

Seaborn is a Python data visualization library based on Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. This article deals with the distribution plots in seaborn which is used for examining univariate and bivariate distributions. In this article we will be discussing 4 types of distribution plots namely:

1 joinplot

2 distplot

3 pairplot

4 rugplot

```python
import seaborn as sns
  


  

df = sns.load_dataset('tips')
  

df.head()
```

Displot

```python
sns.set_style('whitegrid')
sns.distplot(df['total_bill'],color ='red', bins = 30)
```

join plot

```python
sns.jointplot(x ='total_bill', y ='tip', data = df)
```

Pair plot

```python
sns.pairplot(df, hue ="sex", palette ='coolwarm')
```

Categorical Plots

Barplot

```python
sns.set_style('darkgrid')
 

sns.barplot(x ='sex', y ='total_bill', data = df, palette ='plasma')
 
```

Count plot

```python
sns.countplot(x ='sex', data = df)
```

Box plot

```python
sns.boxplot(x ='day', y ='total_bill', data = df, hue ='smoker')
```

violin plot

```python
sns.violinplot(x ='day', y ='total_bill', data = df, hue ='sex')
```

```python
sns.violinplot(x ='day', y ='total_bill', data = df, hue ='sex', split = True)
```

Strip plot

```python
sns.stripplot(x ='day', y ='total_bill', data = df,
			jitter = True, hue ='smoker')
```

```python
sns.stripplot(x ='day', y ='total_bill', data = df,
			jitter = True, hue ='smoker', dodge = True)
```

Factor plot

```python

sns.factorplot(x ='day', y ='total_bill', data = df, kind ='bar')
```
