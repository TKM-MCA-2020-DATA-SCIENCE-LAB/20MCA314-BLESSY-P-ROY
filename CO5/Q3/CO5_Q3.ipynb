```python
#n gram in sentence analysis
import numpy as np
import pandas as pd
from sklearn import model_selection,naive_bayes,svm
```

```python
df=pd.read_csv("/content/sample_data/Reviews.csv",engine='python',error_bad_lines=False)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Score'].value_counts()
```

```python
df['positively rated']=np.where(df['Score']>3,1,0)
```

```python
df.head()
```

```python
df['positively rated'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df["Text"],df["positively rated"],random_state=8)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vect=CountVectorizer(min_df=5,ngram_range=(1,2),).fit(x_train)#min word count should 5 other discrd
```

```python
len(vect.get_feature_names())
```

```python
v=vect.get_feature_names()
```

```python
v
```

```python
x_train_vectorized=vect.transform(x_train)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train_vectorized,y_train)
```

```python
predictions=model.predict(vect.transform(x_test))
```

```python
print(predictions)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import roc_auc_score    #roc 
print("AUC score is",roc_auc_score(y_test,predictions))
```
